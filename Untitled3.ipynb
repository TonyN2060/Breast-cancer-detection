```python
def image_statistics(image_folder):
    image_sizes = []
    pixel_intensities = []

    for image_name in os.listdir(image_folder):
        image_path = os.path.join(image_folder, image_name)
        img = mpimg.imread(image_path)

        # Image size
        image_sizes.append(img.shape)

        # Pixel intensity statistics (for grayscale images)
        if len(img.shape) == 2:
            pixel_intensities.append(np.mean(img))

    return image_sizes, pixel_intensities

# Calculate and visualize image statistics
normal_sizes, normal_intensities = image_statistics(normal_path)
benign_sizes, benign_intensities = image_statistics(benign_path)
malignant_sizes, malignant_intensities = image_statistics(malignant_path)

# Plot histograms of image sizes and pixel intensities
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.hist([x[0] for x in normal_sizes], bins=20, alpha=0.5, label='Normal')
plt.hist([x[0] for x in benign_sizes], bins=20, alpha=0.5, label='Benign')
plt.hist([x[0] for x in malignant_sizes], bins=20, alpha=0.5, label='Malignant')
plt.xlabel('Image Width')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(2, 2, 2)
plt.hist(normal_intensities, bins=20, alpha=0.5, label='Normal')
plt.hist(benign_intensities, bins=20, alpha=0.5, label='Benign')
plt.hist(malignant_intensities, bins=20, alpha=0.5, label='Malignant')
plt.xlabel('Mean Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python

# Function to display random images from a class
def display_random_images(image_folder, class_name, num_images=10):
    plt.figure(figsize=(12, 6))
    plt.suptitle(f"Random {class_name} Images", fontsize=16)

    for i in range(num_images):
        random_image = random.choice(os.listdir(image_folder))
        image_path = os.path.join(image_folder, random_image)
        img = mpimg.imread(image_path)

        plt.subplot(1, num_images, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Sample {i + 1}")

# Display random images from each class
display_random_images(normal_path, "Normal")
display_random_images(benign_path, "Benign")
display_random_images(malignant_path, "Malignant")
plt.show()
```

```python

def augment_image(image_path, output_folder, num_augmented=5):
    # Load the image
    img = cv2.imread(image_path)

    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Apply data augmentation techniques and save augmented images
    for i in range(num_augmented):
        augmented_img = img.copy()

        # Brightness adjustment
        brightness_factor = random.uniform(0.7, 1.3)
        augmented_img = cv2.convertScaleAbs(augmented_img, alpha=brightness_factor, beta=0)

        # Contrast enhancement
        contrast_factor = random.uniform(0.7, 1.3)
        mean_value = int(cv2.mean(augmented_img)[0])
        augmented_img = cv2.addWeighted(augmented_img, contrast_factor, augmented_img, 0, mean_value * (1 - contrast_factor))

        # Histogram equalization
        if len(augmented_img.shape) == 2:
            augmented_img = cv2.equalizeHist(augmented_img)

        # Save the augmented image
        output_path = os.path.join(output_folder, f"augmented_{i}_{os.path.basename(image_path)}")
        cv2.imwrite(output_path, augmented_img)

# Specify the input and output folders for each class
input_folders = [normal_path, benign_path, malignant_path]
output_folders = ["augmented_normal", "augmented_benign", "augmented_malignant"]

# Number of augmented images per original image
num_augmented_per_image = 5

# Apply data augmentation to each class
for input_folder, output_folder in zip(input_folders, output_folders):
    for image_name in os.listdir(input_folder):
        image_path = os.path.join(input_folder, image_name)
        augment_image(image_path, output_folder, num_augmented=num_augmented_per_image)
```

```python
# Function to display random augmented images from a class
def display_random_augmented_images(augmented_folder, class_name, num_images=10):
    plt.figure(figsize=(12, 6))
    plt.suptitle(f"Random Augmented {class_name} Images", fontsize=16)

    augmented_images = os.listdir(augmented_folder)
    random_augmented_images = random.sample(augmented_images, num_images)

    for i, augmented_image in enumerate(random_augmented_images):
        augmented_image_path = os.path.join(augmented_folder, augmented_image)
        img = mpimg.imread(augmented_image_path)

        plt.subplot(1, num_images, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Sample {i + 1}")

# Specify the folders containing augmented images for each class
augmented_folders = ["augmented_normal", "augmented_benign", "augmented_malignant"]

# Display random augmented images from each class
for augmented_folder, class_name in zip(augmented_folders, ["Normal", "Benign", "Malignant"]):
    display_random_augmented_images(augmented_folder, class_name)

plt.show()
```

```python
# Define a function to check if an image is dark or blank
def is_dark_or_blank(image):
    # You can set a threshold for darkness; for example, if the mean pixel value is less than 20
    return np.mean(image) < 20

# Function to load and display dark or blank images in a given category
def display_dark_or_blank_images(category_path, category_name):
    dark_or_blank_images = []
    total_images = 0  # Count total images in the category
    for img_file in os.listdir(category_path):
        img = cv2.imread(os.path.join(category_path, img_file))
        total_images += 1
        if is_dark_or_blank(img):
            dark_or_blank_images.append(img)
    return dark_or_blank_images, total_images

# Function to visualize images with a specified title
def visualize_images(images, title, count, total_count):
    num_images = len(images)
    if num_images > 0:
        blank_percentage = (num_images / total_count) * 100
        print(f"{count} {title} blank images ({blank_percentage:.2f}% of total {total_count} images):")
        plt.figure(figsize=(12, 4))
        for i in range(num_images):
            plt.subplot(1, num_images, i + 1)
            plt.imshow(images[i])
            plt.title(f"{title} Image {i+1}")
            plt.axis('off')
        plt.show()
    else:
        print(f"No {title} blank images found.")

# Initialize counts
blank_normal_count = 0
blank_benign_count = 0
blank_malignant_count = 0

# Load and display dark or blank images in each category, along with percentages
dark_or_blank_normal, total_normal_images = display_dark_or_blank_images(normal_path, 'Normal')
blank_normal_count = len(dark_or_blank_normal)
visualize_images(dark_or_blank_normal, 'Normal Dark/Blank', blank_normal_count, total_normal_images)

dark_or_blank_benign, total_benign_images = display_dark_or_blank_images(benign_path, 'Benign')
blank_benign_count = len(dark_or_blank_benign)
visualize_images(dark_or_blank_benign, 'Benign Dark/Blank', blank_benign_count, total_benign_images)

dark_or_blank_malignant, total_malignant_images = display_dark_or_blank_images(malignant_path, 'Malignant')
blank_malignant_count = len(dark_or_blank_malignant)
visualize_images(dark_or_blank_malignant, 'Malignant Dark/Blank', blank_malignant_count, total_malignant_images)

# Print counts and percentages of blank images by category
print("Counts and percentages of blank images by category:")
print("Normal:", blank_normal_count, f"({(blank_normal_count / total_normal_images) * 100:.2f}%)")
print("Benign:", blank_benign_count, f"({(blank_benign_count / total_benign_images) * 100:.2f}%)")
print("Malignant:", blank_malignant_count, f"({(blank_malignant_count / total_malignant_images) * 100:.2f}%)")
```
